```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# 1. Load the Dataset
iris_data = load_iris()
# Create a DataFrame
df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
# Add the target column (the species labels)
df['species'] = iris_data.target 

# Map target integers (0, 1, 2) to actual species names for clarity
df['species'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# 2. Display Basic Information
print("--- First 5 rows of the dataset ---")
print(df.head())

print("\n--- Dataset Statistical Description ---")
print(df.describe())

# 3. Visualize Patterns (Pairplot)
# This plots pairwise relationships in the dataset to reveal patterns
sns.pairplot(df, hue='species', markers=["o", "s", "D"])
plt.suptitle("Pairplot of Iris Dataset Features", y=1.02) # Add title above plot
plt.show()
```
